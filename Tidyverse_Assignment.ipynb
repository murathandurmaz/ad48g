```r
library(tidyverse)
library(DT)
```

```r
data(swiss)
```

```r
head(swiss)
```

```r
swiss %>% select(-Fertility, -Agriculture)
```

```r
swiss %>%
filter( Infant.Mortality > 20 & Education <10)
```

```r
swiss %<>%
mutate(educated_agro = (Agriculture + 2) * (Education / 2 )/5) %>%
mutate(agro_per_examine = Examination / educated_agro)
```

```r
swiss
```

```r
swiss %>%
summarise(average_agro = mean(agro_per_examine, na.rm = T))
```

```r
swiss %>%
group_by(Agriculture) %>%
summarise(average_agro = mean(agro_per_examine, na.rm = T))
```

```r
swiss_agro <- swiss %>%
group_by(Agriculture,Education) %>%
summarise(max_agro = max(educated_agro, na.rm = T),
          min_agro = min(educated_agro, na.rm = T))
```

```r
swiss_agro
```

```r
swiss_agro_long <- swiss_agro %>%
  pivot_longer(cols = c(max_agro, min_agro), 
               names_to = "agro_type", 
               values_to = "agro_value")
```

```r
swiss_agro_long
```

```r
swiss_agro_wide <- swiss_agro_long %>%
  pivot_wider(names_from = agro_type, values_from = agro_value)
```

```r
swiss_agro_wide
```
